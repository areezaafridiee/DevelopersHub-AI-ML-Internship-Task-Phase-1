# Task 1: Exploring and Visualizing the Iris Dataset

## Problem Statement

The objective of this task is to learn how to load, inspect, analyze, and visualize a dataset using Python libraries such as Pandas, Matplotlib, and Seaborn. The Iris dataset is used to explore feature distributions, relationships between variables, and potential outliers.

## Dataset Loading

The Iris dataset is loaded using Pandas and stored in a DataFrame for further analysis.

```python
#import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#load the dataset using pandas
iris=pd.read_csv('iris.csv')
#shape
print('Shape:',iris.shape)
#columns names
print('\nColumns Names:\n',iris.columns)
#head
print('\nFirst Few Rows:\n',iris.head())
#info for summary statistics
print('\nDataset Information:', iris.info())
#describe for summary statistics
print('\nSummary Statistics:')
print(iris.describe())
#missing values
print('\nMissing Values:')
print(iris.isnull().sum())
#scatterplot
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Scatterplot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
#histograms to show value distribution for each feature
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.histplot(iris['sepal_length'], bins=15, ax=axes[0,0])
axes[0,0].set_title('Sepal Length')
sns.histplot(iris['sepal_width'], bins=15, ax=axes[0,1])
axes[0,1].set_title('Sepal Width')
sns.histplot(iris['petal_length'], bins=15, ax=axes[1,0])
axes[1,0].set_title('Petal Length')
sns.histplot(iris['petal_width'], bins=15, ax=axes[1,1])
axes[1,1].set_title('Petal Width')
plt.tight_layout()
plt.show()
#boxplots to identify outliers
plt.figure(figsize=(8,6))
sns.boxplot(data=iris)
plt.title('Boxplots of Sepal Length, Sepal Width, Petal Length, and Petal Width')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()

```

## Data Exploration

The dataset shape, column names, first few rows, data types, and summary statistics are examined to understand the structure of the data.

## Scatter Plot

A scatter plot is created to visualize the relationship between sepal length and petal length across different species.

## Histograms

Histograms are used to understand the distribution of each numerical feature in the dataset.

## Boxplots

Boxplots help identify the spread of the data and potential outliers.

## Results and Insights

- The dataset contains 150 observations and 5 columns.
- There are no missing values.
- Petal length and petal width show strong variation among species.
- Setosa appears clearly separated from the other species in the scatter plot.
- A few potential outliers can be observed in sepal width.
